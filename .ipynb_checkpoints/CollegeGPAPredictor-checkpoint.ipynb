```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./sat_gpa_v1.csv')
df
```

```python
print('the number of null values in data is ', df.isna().sum())
print('the number of duplicated values in data is ', df.duplicated().sum())

#given the limited number of varibles I believe it is highly likely that the 49 duplicates 
# can happen naturally in the dataset and are true values. Meaning they should not be dropped out of the data used.
```

```python
#plotting the scatter plot of GPA and SAT score
plt.figure(figsize=(10, 6))
plt.scatter(y= 'fy_gpa', x='sat_sum', data=df)
plt.title('Relationship between GPA and SAT Scores')
plt.ylabel('GPA')
plt.xlabel('SAT Scores')

plt.show()
```

```python
import statsmodels.api as sm
```

```python
x = df['sat_sum']
y = df['fy_gpa']
x = sm.add_constant(x)

# Fit the regression model
model = sm.OLS(y, x).fit()
model
```

```python

# Plot the data points and the regression line
plt.scatter(x='sat_sum',y= 'fy_gpa', data=df, marker='o', linestyle='-', alpha=0.2)
plt.plot(x, model.predict(x))
plt.xlabel('SAT Summary (Math + Verbal)')
plt.ylabel('GPA')
plt.title('Correlation between GPA vs. SAT Scores')
plt.show()

model.summary()
```

My analysis of the correlation of Sat Score to GPA is that there is a fairly strong positive one. I know that the scales of the 2 variable are very different with SAT scores ranging from ~450 to 1400 and GPA being from 0-4.0 but the higher the SAT score generally the higher the first year GPA.
